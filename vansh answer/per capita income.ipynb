### Subject
Using *canada_per_capita_income.csv* file build a regression model and predict the per capita income for canadian citizens in year 2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
data
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(data['year'],data['per capita income (US$)'],color='red',marker='+')
```

```python
year = data[['year']]
year
```

```python
income = data['per capita income (US$)']
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_ # m
```

```python
reg.intercept_ # b
```

```python
# Y = m * X + b
828.46507522*2020 + -1632210.7578554575
```

```python
predicted_income = reg.predict(year)
predicted_income
```

```python
predicted_data = pd.DataFrame(year)
predicted_data.head(3)
```

```python
predicted_data['predicted income'] = predicted_income
predicted_data.head(3)
```

```python
%matplotlib inline
plt.xlabel('year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(data['year'],data['per capita income (US$)'],color='red',marker='+')
plt.plot(predicted_data['year'],predicted_data['predicted income'],color='blue')
```
